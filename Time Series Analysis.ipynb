## Time Series

```python
We must need a feature which should be >> Time

It is a set of observation/ datapoints taken at equal interval of time;
where our main variable should be time(ind var) and time dependent(Target)
```

```python
In TSA:
    Univariate Analysis:
        1. Time/date
        2. Target
        
    Multivariate Analysis:
        1. Time/date
        2. Target
        3. Extra features (Which will help our model to predict output)
```

```python
Types of TSA:
    Univariate Analysis:(1 time feature, 1 time dependent feature)
        AutoRegressive(AR)
        Moving Average(MA)
        AutoRegressive Moving Average (ARMA)
        ARIMA (Integrated)
        SARIMA(Seasonal)
        FBProphet
        LSTM(Long Short Term Memory)
        
        
    Multivariate Analysis:
        VAR(Vector)
        FBProphet
        LSTM
        SARIMAX
```

```python
When we plot data:
    on x axis: time
        y    : magnitude, number, values, amount
```

```python
Limitations:
    Time dependent var is required
    Constant interval of time is required (if datapoints are not in equal interval,
                                          Then we can resample the data)
```

```python
Interval:
    Microsec, sec, min, hour, a day, a week , a month, quarter, year, decade, century
```

```python
How to check Stationary:
    1. Rolling Statistic (visualisation)
    2. ADCF test (Augmented Dickey Fuller Test) (pvalue < 0.05)  >> Data is stationary (Statistical test)
```

```python
H0 : Null Hypothesis : Data is not stationary
Ha : Alternate Hypothesis: Data is stationary
ADCF test comprises of (pvalue, test statistics, critical values)
```

```python
How to make data stationary:
    by creating a new feature:
        1. Differencing ( shift value and take diff)
        2. Rolling mean (window) avg of previous values
        3. Transformations:
            log
            sqrt
            cbrt
            rcpr
            exp
```
